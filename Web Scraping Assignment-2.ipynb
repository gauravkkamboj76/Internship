Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.

This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

Note: All of the above steps have to be done in code. No step is to be done manually.

```python
# Step 1): open automated chrome software for browsing:

driver1 = webdriver.Chrome(r"c:\web driver\chromedriver.exe")

driver1.get("https://www.naukri.com/")

# Step 2): Fill the requirements in chrome designation: "Data Analyst" and location: "Banglore":
field_designation = driver1.find_element_by_class_name('suggestor-input ')
field_designation.send_keys('Data Analyst')

field_location = driver1.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[5]/div/div/div/input')
field_location.send_keys('Banglore')

#  Step 3): Clicked on the search button:
search_button = driver1.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()
```

```python

job1_title = []
experience1_title = []
location1_title = []
company1_title = []

# Step 4): Scrapeing the data:
job_title_tag=driver1.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
for i in job_title_tag:
    jtitle= i.text
    job1_title.append(jtitle)
    
experience_title_tag=driver1.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")
for i in experience_title_tag:
    etitle= i.text
    experience1_title.append(etitle)
    
location_title_tag = driver1.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
for i in location_title_tag:
    ltitle=i.text
    location1_title.append(ltitle)
    
    
company_title_tag = driver1.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
for i in company_title_tag:
    ctitle=i.text
    company1_title.append(ctitle)
    
```

```python
# Step 5): Creating the data Frame: 
naukri=pd.DataFrame({"Job":job1_title, "Experience":experience1_title, "Location":location1_title, "Company":company1_title})
Data_Analyst=naukri[0:10]
Data_Analyst
```

```python
# Step 6): Close the Chrome:
driver1.close()
```

```python
driver10=webdriver.Chrome("chromedriver.exe")
driver10.get("https://www.ambitionbox.com/")
```

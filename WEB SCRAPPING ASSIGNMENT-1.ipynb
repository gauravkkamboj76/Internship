Ques.2:Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.




```python
page2=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
page2
```

```python
soup2=BeautifulSoup(page2.content)
soup2
```

```python
name2=[]
for i in soup2.find_all('h3',class_="lister-item-header"):
    name2.append(i.text[4:-8])

release2=[]
for i in soup2.find_all('span',class_="lister-item-year text-muted unbold"):
    release2.append(i.text[1:5])
    
rating2=[]
for i in soup2.find_all('div',class_="ratings-bar"):
    rating2.append(i.text[3:6])
```

```python
IMDB_Top_rated_100_movies=pd.DataFrame({'Name':name2,'Rating':rating2,'Year of Release':release2})
IMDB_Top_rated_100_movies
```

```python
IMDB_Top_rated_100_movies.info()
```

Ques.9:Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page9=requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=Restaurant')
page9
```

```python
soup9=BeautifulSoup(page9.content)
soup9
```

```python
restaurant9=[]
for i in soup9.find_all("div", class_="restnt-info cursor"):
    restaurant9.append(i.text[:-30])
    
location9=[]
for i in soup9.find_all("div", class_="restnt-loc ellipsis"):
    location9.append(i.text)
    
rating9=[]
for i in soup9.find_all("div",class_="restnt-rating rating-4"):
    rating9.append(i.text)
    
cuisine9=[]
for i in soup9.find_all("span",class_="double-line-ellipsis"):
    cuisine9.append(i.text.split("|")[1]) 
    
image9=[]
for i in soup9.find_all("div",class_="img-wrap"):
    image9.append(i.find("img",class_="no-img").get("data-src"))
```

```python
Dineout=pd.DataFrame({'Restaurant Name':restaurant9,'Cuisine':cuisine9,'Location':location9,'Ratings':rating9,'Image URL':image9})
Dineout
```

```python
Dineout.info()
```

 iv) h5-median

```python
page10=requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page10
```

```python
Detail_of_top_publication=pd.DataFrame({"Rank":rank10,"Publication":publication10,"H5 Index":index10_a,"H5 Median":median10})
Detail_of_top_publication

```
